```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Symptoms_DS.csv")
```

```python
y=pd.get_dummies(df["prognosis"])
y
```

```python
y=y.replace({True: 1, False: 0})
y
```

```python
x=df.drop(["prognosis"],axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val_test, y_train, y_val_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_val, x_test, y_val, y_test = train_test_split(x_val_test, y_val_test, test_size=0.5, random_state=42)
```

```python
x_train.shape,x_test.shape,x_val.shape,y_train.shape,y_test.shape,y_val.shape
```

```python
x_train.head()
```

```python
y_train.head()
```

**Deep Neural Network**

```python
from keras.models import Sequential
from keras.layers import Dense , Dropout
from tensorflow.keras.optimizers import Adam
```

```python
import tensorflow as tf
tf.__version__
```

```python
model=Sequential()
model.add(Dense(units=132,activation="relu",kernel_initializer="random_uniform"))
model.add(Dropout(0.5))
model.add(Dense(units=66,activation="relu",kernel_initializer="random_uniform"))
model.add(Dropout(0.5))
model.add(Dense(units=33,activation="relu",kernel_initializer="random_uniform"))
model.add(Dropout(0.5))
model.add(Dense(units=41,activation="softmax",kernel_initializer="random_uniform"))
```

```python
model.compile(loss="categorical_crossentropy",optimizer=Adam() ,metrics=["accuracy"])
history = model.fit(x_train, y_train, batch_size=32, epochs=50, verbose=2, validation_data=(x_val,y_val)) 
```

```python
model.summary()
```

```python
model.save("epics_model_vsc.h5")
```
